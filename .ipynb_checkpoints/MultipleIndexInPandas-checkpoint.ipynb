```python
# can we have multiple index? Let's try
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
a = pd.Series([1,2,3,4,5,6,7,8],index=index_val)
a
```

```python
#how to create multiple index object 
# 1. pd.MultiIndex.from_tuples()
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
multiindex = pd.MultiIndex.from_tuples(index_val)
multiindex.levels[0]
```

```python
# 2. pd.MultiIndex.from_product()
pd.MultiIndex.from_product([['cse','ese'],[2019,2020,2021,2022]])
```

<h2>Long vs Wide dataframes</h2>

```python
#melt -> simple example branch  --> wide to long 
pd.DataFrame({'cse':[120]}).melt()
```

```python
pd.DataFrame({'cse' : [120],'ece' : [100],'mech' : [50]}).melt(var_name="branch",value_name="num_students")
```

```python
temp_df = pd.DataFrame({
'branch' : ['cse','ece','mech'],
'2020' : [100,150,60],
'2021' : [120,130,80],
'2022' : [150,140,79]
})
temp_df.melt(id_vars=['branch'],var_name ="year" , value_name = "students")
```

```python
#melt --> real world examples 
confirm = pd.read_csv('./time_series_covid19_confirmed_global.csv')
death = pd.read_csv('./time_series_covid19_deaths_global.csv')
```

```python
confirm
```

```python
death
```

```python
death.head()
```

```python
death_melt = death.melt(id_vars=["Province/State","Country/Region","Lat","Long"],var_name="date",value_name="num_deaths")
confirm_melt = confirm.melt(id_vars=["Province/State","Country/Region","Lat","Long"],var_name="date",value_name="num_cases")
confirm_melt.shape
```

```python
death_melt
```

```python
confirm_melt
```

```python
merged_death_confirmation = confirm_melt.merge(death_melt,on=["Province/State","Country/Region","Lat","Long","date"])
```
